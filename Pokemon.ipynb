```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("D:\Some CSVs\Pokemon.csv",index_col=0,encoding="latin-1")
```

```python
data 
```

```python
sns.lmplot(x="Attack",y="Defense",hue= "Stage",data = data,fit_reg=False) #Stage is the evolution of pokemon
sns.lmplot(x="Speed",y="Attack",hue = "Legendary",data = data,fit_reg=False) #Color by legendary
sns.lmplot(x="Speed",y="Defense",hue = "Type 1",data = data,fit_reg=False) #color by Type of pokemon 
```

```python
stats_data= data.drop(["Total","Legendary","Stage"],axis=1)
stats_data.head()
sns.boxplot(data=stats_data)
```
